<h4>Methods<h4>

```python
def total_elements(list):
    count = 0
    for imag in list:
        if imag.startswith("2020"):
            count = count + 1
    return count
```

```python
def slice_per(source, step):
    return [source[i::step] for i in range(step)]
```

<h4>RGB Histogram of all labeled images<h4>

```python
nb_bins = 256
count_r = np.zeros(nb_bins)
count_g = np.zeros(nb_bins)
count_b = np.zeros(nb_bins)

for image in os.listdir("D:/Model/all_labeled_uint16"): # path to labeled tiff
  img = tifffile.imread('D:/Model/all_labeled_uint16/'+image)
  x = np.array(img)
  x = x.transpose(2, 0, 1)
  hist_r = np.histogram(x[0], bins=nb_bins, range=[0, 255])
  hist_g = np.histogram(x[1], bins=nb_bins, range=[0, 255])
  hist_b = np.histogram(x[2], bins=nb_bins, range=[0, 255])
  count_r += hist_r[0]
  count_g += hist_g[0]
  count_b += hist_b[0]

bins = hist_r[1]
fig, ax = plt.subplots(figsize=(15, 10))
plt.bar(bins[:-1], count_r, color='r', alpha=0.7)
plt.bar(bins[:-1], count_g, color='g', alpha=0.7)
plt.bar(bins[:-1], count_b, color='b', alpha=0.7)
plt.xlabel('Pixel Brightness', fontsize = 15)
plt.ylabel('Number of Pixels', fontsize = 15)
plt.yticks(rotation=45)
plt.ticklabel_format(style = 'plain', useLocale = True)
ax.yaxis.set_major_formatter(ticker.EngFormatter())
plt.title('RGB Histogram of all labeled Images', fontweight = 'bold', fontsize = 15)
plt.legend(['Red_Channel', 'Green_Channel', 'Blue_Channel'])
```

<h4>Average RGB Histogram of Flight Year 2020<h4>

```python
images2020 = total_elements(os.listdir("D:/Model/all_labeled_uint16")) # access to def total_elements()
print("There are", images2020, "images from flight year 2020.")
```

```python
nb_bins = 256
count_r = np.zeros(nb_bins)
count_g = np.zeros(nb_bins)
count_b = np.zeros(nb_bins)

for image in os.listdir("D:/Model/all_labeled_uint16"): # path to labeled png
    if image.startswith("2020"):
        img = tifffile.imread('D:/Model/all_labeled_uint16/'+image)
        x = np.array(img)
        x = x.transpose(2, 0, 1)
        hist_r = np.histogram(x[0], bins=nb_bins, range=[0, 255])
        hist_g = np.histogram(x[1], bins=nb_bins, range=[0, 255])
        hist_b = np.histogram(x[2], bins=nb_bins, range=[0, 255])
        count_r += hist_r[0] # dividing through number of images 2020 to get mean values
        count_g += hist_g[0] # dividing through number of images 2020 to get mean values
        count_b += hist_b[0] # dividing through number of images 2020 to get mean values

histr_r = count_r/images2020
histr_g = count_g/images2020
histr_b = count_b/images2020
hist_reference = np.around(np.array((histr_r, histr_g, histr_b), dtype=float))
bins = hist_r[1]
fig, ax = plt.subplots(figsize=(15, 10))
plt.bar(bins[:-1], count_r/images2020, color='r', alpha=0.7)
plt.bar(bins[:-1], count_g/images2020, color='g', alpha=0.7)
plt.bar(bins[:-1], count_b/images2020, color='b', alpha=0.7)
plt.xlabel('Pixel Brightness', fontsize = 15)
plt.ylabel('Number of Pixels', fontsize = 15)
plt.yticks(rotation=45)
plt.ticklabel_format(style = 'plain', useLocale = True)
ax.yaxis.set_major_formatter(ticker.EngFormatter())
plt.title('RGB Histogram from Flight Year 2020 - Average', fontweight = 'bold', fontsize = 15)
plt.legend(['Red_Channel', 'Green_Channel', 'Blue_Channel'])
```

```python
results = []
hist_ref = np.ndarray.flatten(hist_reference) # creating flatten vector
for image in os.listdir("D:/Model/all_labeled_uint16"): # path to labeled png
    if image.startswith("2020"):
        img = tifffile.imread('D:/Model/all_labeled_uint16/'+image)
        x = np.array(img)
        x = x.transpose(2, 0, 1)
        hist_r = np.histogram(x[0], bins=nb_bins, range=[0, 255])
        hist_g = np.histogram(x[1], bins=nb_bins, range=[0, 255])
        hist_b = np.histogram(x[2], bins=nb_bins, range=[0, 255])
        count_r = hist_r[0]
        count_g = hist_g[0]
        count_b = hist_b[0]
        hist_source = np.ndarray.flatten(np.array((count_r, count_g, count_b), dtype=float)) # create matrix and transform to flatten vector for pearson calculation
        result = scipy.stats.pearsonr(hist_ref, hist_source)[0] # calculate pearson correlation
        results.append(result)
        results.append(image)
results = (slice_per(results, 2)) # dividing list in two lists
df = pd.DataFrame(results).transpose().rename(columns = {0: "Correlation", 1: "ImageName"}).sort_values(by="Correlation", ascending=False) # creating df and sort by column "Correlation"
df
```

<h4>Variant 1 - Changing tif uint16 to tif uint8 per channel<h4>

```python
# Description: Changing tif uint16 to tif uint8
# Source: https://pro.arcgis.com/de/pro-app/latest/tool-reference/data-management/copy-raster.htm

import arcpy
import os

rasterpath = "D:/Model/all_labeled_uint16" # path to labeled orthophoto
outFolder = "D:/Model/all_labeled_uint8" # new destination path
arcpy.env.workspace = rasterpath

for rasterFile in arcpy.ListRasters("*.tif"):
    outRaster = os.path.join(outFolder,  rasterFile.replace(".tif", ".tif"))
    arcpy.CopyRaster_management(in_raster=rasterFile,
                                out_rasterdataset=outRaster,
                                pixel_type="8_BIT_UNSIGNED",
                                scale_pixel_value="NONE",
                                format="TIFF")
```

<h4>Variant 2 - Color-infrared images (uint8)<h4>

```python
import arcpy
from arcpy.ia import *

arcpy.CheckOutExtension("ImageAnalyst")

# Set input raster
rasterpath = "D:/Model/all_labeled_uint8" # path to labeled orthophoto
arcpy.env.workspace = rasterpath
outFolder = "D:/Model/all_labeled_uint8_infrared" # new destination path

for rasterFile in arcpy.ListRasters("*.tif"):
    outRaster = os.path.join(outFolder,  rasterFile.replace(".tif", ".tif"))
    rendering = arcpy.ia.Render(in_raster=rasterFile, rendering_rule= {'bands': [4,3,2]})
    arcpy.CopyRaster_management(in_raster=rendering,
                                out_rasterdataset=outRaster,
                                pixel_type="8_BIT_UNSIGNED",
                                scale_pixel_value="NONE",
                                format="TIFF")
```

```python
path_ref = "D:/Model/all_labeled_uint16/2020_2661711.0196_1231867.9929_FID_5521_20cm.tif"
reference = tifffile.imread(path_ref).astype("uint8") # reference image

copy_to_path = "D:/Model/all_labeled_uint8_hm/" # new destination path

for image in os.listdir("D:/Model/all_labeled_uint16"):
        img = tifffile.imread('D:/Model/all_labeled_uint16/'+image).astype("uint8") # source image we want to alter
        img_new = histogram_match(source=img,
                                  reference=reference,
                                  match_proportion=0.7).astype(np.uint8) 
        tifffile.imsave(copy_to_path+image, img_new)
```
