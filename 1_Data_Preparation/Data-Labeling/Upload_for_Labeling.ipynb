Importing modules

```python
import os
import boto3
import pandas as pd
import json
```

AWS Simple Storage Service S3 - properties

```python
ACCESS_KEY = "ACCESS_KEY"
SECRET_KEY = "SECRET_KEY"
bucket = "solarmaster1586"
```

Creating a service client

```python
s3 = boto3.client(service_name='s3',
                  aws_access_key_id=ACCESS_KEY,
                  aws_secret_access_key=SECRET_KEY)
```

Creating paginators

```python
paginator = s3.get_paginator('list_objects_v2')
pages = paginator.paginate(Bucket=bucket)
```

Getting keys (filenames)

```python
keys = []
for page in pages:
  for obj in page['Contents']:
    keys.append(obj['Key'])
```

Getting keys (Object URL)

```python
urls = []
for key in keys:
  urls.append({'image': {'url': f"https://{bucket}.s3.eu-central-1.amazonaws.com/{key}"}})
```

Creating data frame

```python
dataset_identifier = "charlie4611-/solar_panels" # segments.ai
data = {'Name': keys,
        'Attributes': urls}
df = pd.DataFrame(data, columns = ["Name", "Attributes"]).assign(data_id=dataset_identifier)
```
